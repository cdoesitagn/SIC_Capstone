```python
from vnstock3 import Vnstock
import pandas as pd
```

```python
tickers = ['BCM', 'BID', 'BVH', 'CTG', 'FPT', 'GAS', 'GVR', 'HDB', 'HPG', 'MBB', 'MSN', 'MWG', 'PLX', 'POW', 'SAB', 'SHB', 'SSB', 'SSI', 'STB', 'TCB', 'TPB', 'VCB', 'VHM', 'VIB', 'VIC', 'VJC', 'VNM', 'VPB', 'VRE']
```

```python
res = []
cols = None
for ticker in tickers:
    company = Vnstock().stock(symbol=ticker, source='TCBS').company
    comp_overv = company.overview()
    comp_prof = company.profile()
    comp = pd.concat((comp_overv, comp_prof))
    cols = comp.columns
    res.append(comp.values[0])
res
```

```python
res = pd.DataFrame(res, columns=cols)
res
```

```python
final = res.loc[:, ['short_name', 'industry', 'website', 'stock_rating', 'industry_id']]
final
```
